Questão 1 \
Utilizando o dataset 'tips' crie um gráfico de barras mostrando o faturamento total do restaurante por dia, exibindo o total gasto por homens e mulheres. \
Dica: use crosstab.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
summary = pd.crosstab(df['day'], df['sex'], values=df['total_bill'], aggfunc='sum')
summary
```

```python
summary.plot(kind='bar', stacked=False)
plt.title('Faturamento total do restaurante')
plt.ylabel('Valor total da conta')
plt.xlabel('Dia da semana')
plt.show()
```

Questão 2 \
Elabre um gráfico de pizza que mostre a divisão do faturamento entre homens e mulheres.

```python
df_male = df[df['sex'] == 'Male']
df_female = df[df['sex'] == 'Female']

sizes = [df_male['total_bill'].sum(), df_female['total_bill'].sum()]
labels = ['Homens', 'Mulheres']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Divisão de faturamento entre homens e mulheres')
plt.axis('equal')
plt.show()
```

Questão 3 \
Elabore um gráfico de dispersão que mostre a relação entre o número de pessoas por mesa e o valor total da conta.

```python
sns.regplot(x='size', y='total_bill', fit_reg=True, data=df)
plt.title('Relação entre o número de pessoas e o valor da conta')
plt.xlabel('Número de pessoas')
plt.ylabel('Valor da conta')
plt.show()
```
