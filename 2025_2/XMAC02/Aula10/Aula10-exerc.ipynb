1) Exiba histogramas lado a lado mostrando o índice de corrupção na Europa e na África. Use bins=15.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CPI_2014.csv', encoding='latin-1')

dfEurope = df[df['Continent'] == 'Europe']
dfAfrica = df[df['Continent'] == 'Africa']
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = [8, 6]

plt.subplot(1, 2, 1)
sns.histplot(dfEurope['CPI_2014'], kde=False)
plt.title('Europa')
plt.xlabel('Nível de corrupção')
plt.axvline(60, color='k', linestyle='--')

plt.subplot(1, 2, 2)
sns.histplot(dfAfrica['CPI_2014'], kde=False)
plt.title('África')
plt.xlabel('Nível de corrupção')
plt.axvline(60, color='k', linestyle='--')

plt.tight_layout()
plt.show()
```
