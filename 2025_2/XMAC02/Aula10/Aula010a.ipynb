## Gráfico de Linhas
Vamos criar um gráfico de linhas para analisar a evolução do preço das ações das empresas Tesla e Apple.

```python
# pip install yfinance
# pip install --upgrade yfinance
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ticker = yf.Ticker('TSLA')
df1 = ticker.history(interval='1d', start='2020-01-01')
df1.head()
```

```python
ticker = yf.Ticker('AAPL')
df2 = ticker.history(interval='1d', start='2020-01-01')
df2.head()
```

```python
ticker = yf.Ticker('NVDA')
df3 = ticker.history(interval='1d', start='2020-01-01')
df3.head()
```

```python
df1['Close AAPL'] = df2['Close']
df1['Close NVDA'] = df3['Close']

df1.rename(columns={'Close': 'Close TSLA'}, inplace=True)
df1.head()
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
sns.set_style('darkgrid')

df1[['Close TSLA', 'Close AAPL', 'Close NVDA']].plot()
plt.title('Valor das ações')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.show()
```

Vamos agora utilizar o dataset LifeExpectancy para criar um gráfico de barras.

```python
df4 = pd.read_csv('lifeExpectancy.csv')
df4.head()
```

Nosso gráfico deve permitir comparar a expectativa de vida de homens e mulheres de países africanos.

```python
df4.plot(x='Countries', kind='bar', stacked=False)
plt.title('Expectativa de vida de homens e mulheres na África')
plt.xlabel('Países da África')
plt.ylabel('Expectativa de vida')
plt.show()
```

Para uma melhor visualização, podemos dividir o dataset. A seguir vamos tomar os 15 primeiros países do dataset.

```python
df5 = df4.head(15)

df5.plot(x='Countries', kind='bar', stacked=False)
plt.title('Expectativa de vida de homens e mulheres na África')
plt.xlabel('Países da África')
plt.ylabel('Expectativa de vida')
plt.show()
```
