### Aula 24 - Equações Lineares & Aula 25 - Resolução de Sistemas Lineares

```python
import numpy as np

A = np.array([[8, 25, 16],
             [25, 87, 55],
             [16, 55, 36]])

b = np.array([397_000, 1_281_100, 817_700])

solution = np.linalg.solve(A, b)

print(f'Beta0 = {solution[0]:9,.2f}')
print(f'Beta1 = {solution[1]:9,.2f}')
print(f'Beta2 = {solution[2]:9,.2f}')
```
